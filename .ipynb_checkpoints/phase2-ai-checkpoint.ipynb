```python
import tensorflow as tf
```

```python
from tensorflow import keras
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
from keras.datasets import cifar10
```

```python
(x_train, y_train),(x_test, y_test) = cifar10.load_data()
```

```python
for i in range(10):
    ax = plt.imshow(x_train[i])
    plt.title(y_train[i])
    plt.show()
```

```python
#Converting y_train and y_test to one hot encoding for potentially better accuracy
y_train_one_hot = keras.utils.to_categorical(y_train, 10)
y_test_one_hot = keras.utils.to_categorical(y_test, 10)
```

```python
print('The one hot label is:', y_train_one_hot[0])
```

```python
#Make x_train and x_test values between 0-1 so that model can work better
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train = x_train/255
x_test = x_test/255
```

```python
x_train[0]
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
```

```python
#Added three convolutional layers and two fully connected layers, as well as some maxpooling and
#dropout layers in between
model = Sequential()
```

```python
model.add(Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(32,32,3)))
```

```python
model.add(Conv2D(32, (3, 3), activation='relu', padding='same'))
```

```python
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
```

```python
model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(512, activation='relu'))
```

```python
model.add(Dropout(0.5))
```

```python
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam', loss='categorical_crossentropy', metrics = ['accuracy'])
```

```python
#First trained for 10 epochs
training = model.fit(x_train, y_train_one_hot, batch_size=32, epochs=10, validation_split=0.2)
```

```python
#Evaluate model, validation accuracy of 0.75, as the validation loss hasn't arrived at a plateau
#yet, will try to see if accuracy would increase after another 10 epochs
model.evaluate(x_test, y_test_one_hot)
```

```python
loss = training.history['loss']
val_loss = training.history['val_loss']
```

```python
import pandas as pd
```

```python
ax = pd.DataFrame(loss).plot()
pd.DataFrame(val_loss).plot(ax=ax)
ax.legend(["loss","val_loss"])
```

```python
model.save('cifar10_model.h5')
```

```python
prediction = model.predict(x_test[0:1])
prediction
import numpy as np
np.argmax(prediction)
```

```python
y_pred = model.predict(x_test)
y_pred = np.argmax(y_pred,axis=1)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
import seaborn as sns
sns.heatmap(confusion_matrix(y_test, y_pred),annot=True)
```

```python
training2 = model.fit(x_train, y_train_one_hot, batch_size=32, epochs=10, validation_split=0.2)
```

```python
#Evaluate model after another 10 epochs, validation accuracy has increased to 0.77
model.evaluate(x_test, y_test_one_hot)
```

```python
loss2 = training2.history['loss']
val_loss2 = training2.history['val_loss']
```

```python
#Validation loss has not started increasing yet, but seems to have stopped the downward trend, model
#is probably not overfit
ax = pd.DataFrame(loss2).plot()
pd.DataFrame(val_loss2).plot(ax=ax)
ax.legend(["loss","val_loss"])
```

```python
np.argmax(prediction)
y_pred2 = model.predict(x_test)
y_pred2 = np.argmax(y_pred2,axis=1)
y_pred2
```

```python
confusion_matrix(y_test, y_pred2)
sns.heatmap(confusion_matrix(y_test, y_pred2),annot=True)
```

```python
#Train accuracy slope is steeper than validation accuracy now, which might be an indication that
#the model is getting close to overfitting
acc2 = training2.history['acc']
val_acc2 = training2.history['val_acc']
ax = pd.DataFrame(acc2).plot()
pd.DataFrame(val_acc2).plot(ax=ax)
ax.legend(["acc","val_acc"])
```
